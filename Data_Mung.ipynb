Missing Values

```
df.isna().sum()
```

We can drop those rows which consist missing values or we can calculate the mean, min, max and median .

fill in the missing values 

```
age_mean_value=df['Age'].mean()
df['Age']=df['Age'].fillna(age_mean_value)
```

```
df
```

Remove coulmn

```
df.drop("Cabin",axis=1,inplace=True)
```

```
df
```

Filtering data

```
filtered_age = df[df.Age>45]
filtered_age
```

Sorting data

```
# let's sort the column Name in ascending order
sorted_passengers = df.sort_values('Name',ascending=True,kind ='heapsort')
# now print the sorted data
sorted_passengers.head()
```

Merging

```
merged_df = pd.merge(df.head(),df.tail(),how='outer',indicator=True)
merged_df
```
